1º passo: Ler arquivo excel com pandas

```
import pandas as pd

tabela_completa = pd.read_excel("/content/drive/MyDrive/Estudos/Python/Intensivo Lira Hashtag/Aula 01/Vendas.xlsx")

display(tabela_completa)
```

2º passo: Calcular faturamento das lojas

```
tabela_faturamento = tabela_completa[["ID Loja","Valor Final"]].groupby("ID Loja").sum()

display(tabela_faturamento)
```

3º passo: Calcular qtde de produtos vendidos por loja

```
tabela_qtde = tabela_completa[["ID Loja","Quantidade"]].groupby("ID Loja").sum()

display(tabela_faturamento)
```

4º passo: Calcular Tiket médio por loja

```
tabela_tiket_medio = (tabela_faturamento["Valor Final"] / tabela_qtde["Quantidade"]).round(2)
tabela_tiket_medio = tabela_tiket_medio.to_frame()
tabela_tiket_medio.rename(columns={0:"Tiket médio"}, inplace = True)
display(tabela_tiket_medio)
```
